```
muestra1.media.N=function(Conf,sigma,e,N){
   alfa=(1-Conf)
  z=qnorm(Conf+alfa/2)
  n0=z^2*sigma/e^2
  n1=n0/(1+(n0-1)/N)
  return(round(n1))

}
```

```
muestra1.media.N(0.95,var(alcohol),0.05, length(alcohol))
```

```
ic.m.csigma<-function(media,desv_st,n,ic=95){
    Error_st <- desv_st/sqrt(n)
  ci.y <- 1-(((100-ic)/100)/2)
  Error <- qnorm(ci.y)*Error_st
  Lim.Inf <- Media - Error
  Lim.Sup <- Media + Error
  result <- data.frame(Media, desv_st, n,Error_st, 
                       Error, Lim.Inf, Lim.Sup)
  return(result)

}
##ic.m.csigma(mean(vinosa$alcohol),sd(vinosa$alcohol),length(vinosa$alcohol),ic=95)

```
